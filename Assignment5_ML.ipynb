```python
### Your code here
def secondBest(probability_list):
  second_best_pred_list = []
  for i in probability_list:
    sorted_probabliity_list = np.sort(i)
    second_best_pred = sorted_probabliity_list[-2]
    result = np.where(i == second_best_pred)
    second_best_pred_list.append(int(result[0]))
    
  #print(second_best_pred_list)
  return(second_best_pred_list)

first_10 = probability_predictions[0:10]
#secondBest(probability_predictions)
secondBest(first_10)
```

## <font color='blue'>Q1: Working with Warm Start to do Focused Training

<font color='black'>

#### 1. What is warm start?
Looking closely into the setup of scikit-learn MLPClassifier, you will see an option 'warm-start'. Putting warm start in simple terms:

cold_mode = classifier(warm_start = False) <br>
cold_model.fit(x,y) <br>
print(cold_model.coefs_, cold_model.intercepts_) <br>
--> Result 0.123, 0.345 <br>
cold_model.fit(x,y) <br>
print(cold_model.coefs_, cold_model.intercepts_) <br>
--> Result 0.333, 0.234 <br>
====> After the 2nd run of .fit(), values are completely different. <br>

warm_model = classifer(warm_start = True) <br>
warm_model.fit(x,y)<br>
print(warm_model.coeffs_, warm_model.intercepts_) <br>
--> Result 0.235, 0.123 <br>
warm_model.fit(x,y) <br>
print(warm_model.coeffs_, warm_model.intercepts_) <br>
--> result 0.235, 0.123<br>
====> Parameters changes relatively little because it starts with the set of weights (i.e. parameters) of the first fit() call.



** Note that you have to make sure that the unique classes during every "fit" call must be the same. You can't fit warm_model.fit(x1,y1) where y1 has 3 unique classes 1,2,3 and call warm_model.fit(x2,y2) where y2 has unique values 2,3 or 2,3,4 (without class 1)
